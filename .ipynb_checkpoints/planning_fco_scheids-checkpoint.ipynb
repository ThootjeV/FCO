```python
import numpy as np
import pandas as pd
from odf import chart, style, table, text
from odf.opendocument import load
from pandas_ods_reader import read_ods
```

```python
path = "doc_FCO.ods"
sheet_idx = 1
copy_w04 = read_ods(path, sheet_idx)
```

```python
rows_with_nan = []

for index, row in copy_w04.iterrows():

    is_nan_series = row.isnull()

    if is_nan_series.all():

        rows_with_nan.append(index)


print(rows_with_nan)
```
